# Best Fit Line for a very large DataSet

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
%matplotlib inline
```

```python
x_data=np.linspace(0.0,10.0,1000000)
noise=np.random.randn(len(x_data))
y_data=0.5*x_data+5+noise
```

### Here I have created a very large dataset of 1 million samples and added a little noise to it. We have created a y=0.5x+5 line with noise

```python
x_df=pd.DataFrame(data=x_data,columns=['X Data'])
y_df=pd.DataFrame(data=y_data,columns=['Y Data'])
my_data=pd.concat([x_df,y_df],axis=1)
```

```python
my_data.sample(250).plot(kind='scatter',x='X Data', y='Y Data')
```

```python
batch_size=8
rand=np.random.randn(2)
m=tf.Variable(rand[0])
b=tf.Variable(rand[1])
```

```python
xph=tf.placeholder(tf.float64,batch_size)
yph=tf.placeholder(tf.float64,batch_size)
```

```python
y_model=m*xph+b
error=tf.reduce_sum(tf.square(y_model-yph))
optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.001)
train=optimizer.minimize(error)
```

```python
init=tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    batches=5000
    for i in range(batches):
        rand_ind=np.random.randint(len(x_data),size=batch_size)
        feed={xph:x_data[rand_ind],yph:y_data[rand_ind]}
        sess.run(train,feed_dict=feed)
    m_pred,b_pred=sess.run([m,b])
```

#### Vary the batches to predict a good result. Here 5000 gave a better result as compared to others.

```python
m_pred
```

```python
b_pred
```

```python
y_pred=m_pred*x_data+b_pred
my_data.sample(250).plot(kind='scatter',x='X Data', y='Y Data')
plt.plot(x_data,y_pred,c='red')
```

# Doing the above task with the help of TF Estimator

```python
feat_cols=[tf.feature_column.numeric_column('x',shape=[1])]
```

```python
estimator=tf.estimator.LinearRegressor(feature_columns=feat_cols)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_eval,y_train,y_eval=train_test_split(x_data,y_data,test_size=0.3,random_state=101)
```

```python
input_func=tf.estimator.inputs.numpy_input_fn({'x':x_train},y_train,
                        batch_size=8,shuffle=True,num_epochs=None)
```

```python
train_input_func=tf.estimator.inputs.numpy_input_fn({'x':x_train},
            y_train,batch_size=8,shuffle=False,num_epochs=1000)
```

```python
test_input_func=tf.estimator.inputs.numpy_input_fn({'x':x_eval},
            y_eval,batch_size=8,shuffle=False,num_epochs=1000)
```

```python
estimator.train(input_func,steps=1000)
```

```python
train_metrics=estimator.evaluate(input_fn=train_input_func,steps=1000)
```

```python
eval_metrics=estimator.evaluate(input_fn=test_input_func,steps=1000)
```

```python
train_metrics
```

```python
eval_metrics
```

```python
brand_new_data=np.linspace(0,10,10)
```

```python
input_fn_predict=tf.estimator.inputs.numpy_input_fn({'x':brand_new_data},shuffle=False)
```

```python
list(estimator.predict(input_fn=input_fn_predict))
```

```python
predictions=[]
for pred in list(estimator.predict(input_fn=input_fn_predict)):
    predictions.append(pred['predictions'])
```

```python
my_data.sample(250).plot(kind='scatter',x='X Data', y='Y Data')
plt.plot(brand_new_data,predictions,'r')
```
