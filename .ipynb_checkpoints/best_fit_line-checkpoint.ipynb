# Regression Example

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
np.random.seed(101)
tf.set_random_seed(101)
```

```python
x_org=np.linspace(0,10,10)
y_org=np.linspace(0,10,10)
```

### We will add some noise to our data

```python
x_data=np.linspace(0,10,10)+np.random.uniform(-1.5,1.5,10)
y_label=np.linspace(0,10,10)+np.random.uniform(-1.5,1.5,10)
```

```python
plt.scatter(x_data,y_label)
plt.plot(x_org,y_org,c='red')
```

#### Here Red line is the exact line, the best fit line(actual result) and blue dots is the dataset, points we are given. With these points we will draw best fit line

```python
rand=np.random.rand(2)
```

```python
m=tf.Variable(rand[0])
b=tf.Variable(rand[1])
```

```python
error=0
for x,y in zip(x_data,y_label):
    y_hat=m*x+b
    error+=(y-y_hat)**2
```

```python
optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.001)
train=optimizer.minimize(error)
```

```python
init=tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    training_steps=100000
    for i in range(training_steps):
        sess.run(train)
    final_slope, final_intercept=sess.run([m,b])
```

##### For more accuracy we shall increase the number of steps. Till here learning steps were 100000 but to save time we will now take values of m and b as the final_slope and intercept values as they are more accurate

```python
x=np.linspace(0,10,1000)
y_ac1=final_slope*x+final_intercept
```

```python
plt.scatter(x_data,y_label)
plt.plot(x_org,y_org,c='red')
plt.plot(x,y_ac1,c='green')
```

```python
m=tf.Variable(final_slope)
b=tf.Variable(final_intercept)
optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.001)
train=optimizer.minimize(error)
init=tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    training_steps=1000
    for i in range(training_steps):
        sess.run(train)
    final_slope, final_intercept=sess.run([m,b])
y_ac2=final_slope*x+final_intercept
```

```python
plt.scatter(x_data,y_label)
plt.plot(x_org,y_org,c='red')
plt.plot(x,y_ac1,c='green')
plt.plot(x,y_ac2,c='yellow')
```
